### Feature Engineering

```python
""" 
    First, we need to calculate the number of singles a batter has produced.
    This metric is not provided but needed for further metric calculations.

    Singles = (Total Hits) - (2B) - (3B) - (HR)
"""

df2['1B'] = df2['H'] - (df2['2B']) - (df2['3B']) - (df2['HR'])
df2
```

```python
df2.head().T
```

```python
""" Total bases refer to the number of bases gained by a batter through his hits. 

    A batter records one total base for a single, two total bases for a double, 
    three total bases for a triple and four total bases for a home run.

    Total bases are used to determine a player's slugging percentage -- 
    which is total bases divided by at-bats. A player can only add
    to total-bases tally through a hit..
    
    Formula, MLB.com
    TB = singles + (2 x doubles) + (3 x triples) + (4 x home runs)


"""
df2['TB'] = df2['1B'] + (2 * df2['2B']) + (3 * df2['3B']) + (4 * df2['HR'])

df2
```

```python
""" Runs Created estimates a player's offensive contribution in terms of total runs. 
    It combines a player's ability to get on base with his ability to hit for extra bases.
    
    Formula, MLB.com: 
    TB x (H + BB) / (AB + BB)"""

df2['RC'] = ( df2['TB'] * (df2['H'] + df2['BB']) ) / (df2['AB'] + df2['BB'])
df2['RC'] = df2['RC'].round()
df2
```

```python
""" Batting average is a player's hits by his total at-bats,
    for a number between zero (shown as .000) and one (1.000)."""
    
df2['B_AVG'] = df2['H'] / df2['AB']
df2['B_AVG']= df2['B_AVG'].round(decimals=3)
df2
```

```python
""" On-Base Percentage forumates how often a player reaches base.

    Equation via baseball-reference:
    OBP = (Hits + Walks + Hit by Pitch) / (At Bats + Walks + Hit by Pitch + Sacrifice Flies)"""

df2['OBP'] = (df2['H'] + df2['BB'] + df2['HBP']) / (df2['AB'] + df2['BB'] + df2['HBP'] + df2['SF'])
df2['OBP'] = df2['OBP'].round(decimals=3)
df2
```

```python
""" Slugging percentage represents the total number of bases a player records per at-bat. 
    Unlike on-base percentage, slugging percentage deals only with hits,
    it does not include walks and hit-by-pitches in its equation
    
    Slugging percentage differs from batting average in that all hits are not valued equally. 
    While batting average is calculated by dividing the total number of hits by the total number of at-bats, 
    the formula for slugging percentage is: (1B + 2Bx2 + 3Bx3 + HRx4)/AB
"""

df2['SLG'] = ((df2['1B']) + (df2['2B'] * 2) + (df2['3B'] * 3) + (df2['HR'] * 4)) / df2['AB']

# use this to check work
# df2.drop(columns=['SLG2'], inplace=True)

df2['SLG'] = df2['SLG'].round(decimals=3)
# df2['SLG2'] = df2['SLG2'].round(decimals=3)
df2
```

```python
""" On-Base Plus Slugging
"""
df2['OPS'] = df2['OBP'] + df2['SLG']
df2
```

```python
""" Isolated Power : measures raw power as by counting multiple base hits. 
    It is typically used to compare two batters with similar batting averages.
    
        Slugging percentage - Batting average
        or
        (1x2B + 2x3B + 3xHR) / At-bats 
     
"""

df2['ISO'] = df2['SLG'] - df2['B_AVG']

df2['ISO_XBH'] = ((1 * df2['2B']) + (2 * df2['3B']) + (3 * df2['HR'])) / (df2['AB'])

df2['ISO_XBH'] = df2['ISO_XBH'].round(decimals=3)
df2
```

```python
""" BABIP measures a player's batting average exclusively on balls hit into the field of play, 
    removing outcomes not affected by the opposing defense (namely home runs and strikeouts).

    For example, a hitter who goes 2-for-5 with a home run and a strikeout would have a .333 BABIP. 
    He's 1-for-3 on the balls he put in play.


    BABIP can be used to provide some context when evaluating both pitchers and hitters. The league average BABIP is typically around .300. 
    Pitchers who have allowed a high percentage of hits on balls in play will typically regress to the mean, and vice versa. 

    In other words, over time, they'll see fewer (or more) balls in play fall for hits, 
    and therefore experience better (or worse) results in terms of run prevention. 
    The same applies for batters who have seen a high or low percentage of their balls in play drop in for hits.

    BABIP = (H - HR)/(AB - K - HR + SF)

"""

# Total Hits minus Home Runs over the sum of At Bats minus Strike Outs and Home Runs + Sacrefice Flies
df2['BABIP'] = (df2['H'] - df2['HR']) / (df2['AB'] - df2['SO'] - df2['HR'] + df2['SF'])
df2['BABIP'] = df2['BABIP'].round(decimals=3)
df2
```

### Creating a baseline model

```python
def train_linear_regression (X, y):

    """
    Overall, this function implements the closed-form solution for linear regression
    by computing the weight vector using the normal equations. 

    Specifically, the function computes the weight vector w_full that minimizes the sum of squared errors
    between the predicted values of y (based on X and w_full) and the actual values of y.
    """
    # The function starts by creating a column vector of ones,
    # with the number of rows equal to the number of rows in X. 
    # This vector is stored in a variable called ones.
    ones = np.ones(X.shape[0])
    # The X matrix is then augmented by adding a column of ones to the left of X. 
    # This is done using the NumPy column_stack function to horizontally stack ones and X. 
    # The resulting matrix is assigned back to X:
    X = np.column_stack([ones, X])

    #  transpose the matrix above to find Gram matrix
    XTX = X.T.dot(X)
    # now inverse the Gram matrix 
    XTX_inv = np.linalg.inv(XTX)
    # computes the weight vector w_full by taking the matrix product of XTX_inv with the transpose of X, and then with y.
    w_full = XTX_inv.dot(X.T).dot(y)

    # Finally, the function returns the intercept term of w_full 
    # (which corresponds to the first element of w_full) 
    # and the coefficients of w_full (which correspond to the remaining elements of w_full).
    return w_full[0], w_full[1:]
```

### Root Squared Mean Error

RSME is the average difference between each prediction in relation to each target

    RSME = SQRT (
        AVG ( (g(Xi) - Yi)^2) ) 
        )


```python
"""
    This function evaluates the performance of regression models and comparing different models. 
    
    Lower values of RMSE indicate better performance of the model. 
    By computing the RMSE between the predicted and actual values, 
    we can get a sense of how well the model is able to predict the actual values.
"""

def rmse(y_train, y_pred):
    # compute error of two arrays, y_train and y_pred, representing the actual and predicted values respectively
    error = y_pred - y_train

    # compute the squared error (SE) between the predicted and actual values by squaring the error computed in the previous step
    se = error ** 2
    # computes the mean squared error (MSE) by taking the average of the squared errors across all the examples
    mse = se.mean()
    # computes the square root of the mean squared error to get the RMSE, which is a commonly used metric for evaluating the performance of regression models.
    return np.sqrt(mse)
```

### Validating the model

```python
# baseline features, no feature engineering
base = [
    'G', 
    'AB', 
    'R', 
    'H', 
    '2B',
    '3B', 
    'HR', 
    'RBI', 
    'SB', 
    'CS', 
    'BB', 
    'SO', 
    'IBB', 
    'HBP', 
    'SH', 
    'SF', 
    'GIDP']

# filling with 0 tells model to ignore the feature as 0 multiplied by weight equals 0
X_train = df_train[base].fillna(0).values

w0, w = train_linear_regression(X_train, y_train)

#  bias charm + train matrix(weight matrix)
y_pred = w0 + X_train.dot(w)
```

```python
# creating a function that prepares dataframes uniformly 

def prepare_X(df):
    df_num = df[base]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X
```

```python
# Prepare the Training dataframe, then train the linear model, creating weights + bias charm
X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)

# Prepare the Validation dataframe, create the predictions based off Validation df
X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)
```
